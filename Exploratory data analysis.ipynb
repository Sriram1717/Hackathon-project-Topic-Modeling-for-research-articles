```python
# looking at correlation between articles of different classes
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(),annot=True,cmap='coolwarm')
```

```python
# list of categories
list_1 = ['Computer Science', 'Physics', 'Mathematics',
       'Statistics', 'Quantitative Biology', 'Quantitative Finance']
```

```python
# showing no. of instances of each category 
for col in list_1:
    sns.countplot(x=col,data=data)
    plt.show()
```

```python
# let's check if length of abstract gives any indication on the article's category
data['ABSTRACT_length'] = data['ABSTRACT'].apply(lambda x: len(x))
```

```python
data.corr()['ABSTRACT_length']
```

```python
for i in list_1:
    sns.boxplot(x=i,y='ABSTRACT_length',data=data)
    plt.show()
```
